```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0, 0], [1, 0], ':')
plt.title('Sigmoid Function')
plt.show()
```
